```python
import pandas as pd
import numpy as np

df1 = pd.DataFrame(np.random.randint(25, size=(4, 4)), 
				index=["1", "2", "3", "4"], 
				columns=["A", "B", "C", "D"])

df2 = pd.DataFrame(np.random.randint(25, size=(6, 4)),
				index=["5", "6", "7", "8", "9", "10"], 
				columns=["A", "B", "C", "D"])

df3 = pd.DataFrame(np.random.randint(25, size=(4, 4)),
				columns=["A", "B", "C", "D"])

df4 = pd.DataFrame(np.random.randint(25, size=(4, 4)),
				columns=["E", "F", "G", "H"])

display(df1, df2, df3, df4)
```

```python
vert_concate=pd.concat([df1,df2], axis=1)
vert_concate
```

```python
hor_concate=pd.concat([df1,df2], axis=0)
hor_concate
```

## Merge --- ( Join )

```python
dummy_data1 = {
        'id': ['1', '2', '3', '4', '5'],
        'feat1': ['A', 'C', 'E', 'G', 'I'],
        'feat2': ['B', 'D', 'F', 'H', 'J']}

dummy_data2 = {
        'id': ['1', '2', '6', '7', '8'],
        'feat1': ['K', 'M', 'O', 'Q', 'S'],
        'feat2': ['L', 'N', 'P', 'R', 'T']}

df1 = pd.DataFrame(dummy_data1, columns = ['id', 'feat1', 'feat2'])
df2 = pd.DataFrame(dummy_data2, columns = ['id', 'feat1', 'feat2'])

display(df1, df2)
```

```python
df_join = pd.merge(df1, df2, on='id', how='inner')

df_join 
```

```python
df_left = df1.merge(df2, on='id', how='left')

df_left
```

```python
df_right = df1.merge(df2, on='id', how='right')

df_right
```
